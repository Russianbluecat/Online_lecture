```python
# 타이틀만 가져와서 출력하기 

import requests
from bs4 import BeautifulSoup

res  = requests.get('https://davelee-fun.github.io/trial/board/news.html')
if res.status_code ==200 : 

    soup = BeautifulSoup(res.content,'html.parser')

    items = soup.select('div.list_item')

    for item in items : 
        title = item.select_one('span.subject_fixed')
        #냅다 print(title.get_text().strip()) 실행하면 오류남 
        # 왜냐하면 list_item 요소들 중 subject_fixed 가 없는 부분도 있어 그래서 if title: 이라고 명령해야 함 
        if title : 
            print(title.get_text().strip())
else:
    print(f"Failed to retrieve the page. Status code: {res.status_code}")
```

```python
#타이틀 번호 추가하기
#화면상에 보이는 글 제목 왼쪽 숫자 아님 . 그냥 글 순서대로 1,2,3,4, ... 이렇게 숫자를 붙이는 것임 

import requests
from bs4 import BeautifulSoup

res  = requests.get('https://davelee-fun.github.io/trial/board/news.html')
if res.status_code ==200 : 

    soup = BeautifulSoup(res.content,'html.parser')

    items = soup.select('div.list_item')

    index = 0 

    for item in items : 
        #냅다 for index, item in enumerate(items) 라고 시작하면 오류남 
        title = item.select_one('span.subject_fixed')
        if title : 
            index += 1
            # 인덱스를 2자리로 고정 (최대 자릿수만큼 0을 채우거나 공백을 채움)
            print(f"{str(index).rjust(3)}   {title.get_text().strip()}")
else:
    print(f"Failed to retrieve the page. Status code: {res.status_code}")
```

```python
#타이틀별 댓글 수 추가하기

import requests
from bs4 import BeautifulSoup

res  = requests.get('https://davelee-fun.github.io/trial/board/news.html')
if res.status_code ==200 : 

    soup = BeautifulSoup(res.content,'html.parser')

    items = soup.select('div.list_item')

    index = 0 

    for item in items : 
        #냅다 for index, item in enumerate(items) 라고 시작하면 오류남 
        title = item.select_one('span.subject_fixed')
        replies = item.select_one('span.rSymph05')
        
        if title : 
            index += 1
            # 인덱스를 3자리로 고정 (최대 자릿수만큼 0을 채우거나 공백을 채움)
            print(f"{str(index).rjust(3)}   {title.get_text().strip()} [{replies.get_text()}]")
else:
    print(f"Failed to retrieve the page. Status code: {res.status_code}")
```

```python
#엑셀로 저장하기 
#한 행에 글 제목 + 댓글 수 
#그 다음 행은 댓글 

import requests
from bs4 import BeautifulSoup
import openpyxl

res  = requests.get('https://davelee-fun.github.io/trial/board/news.html')
if res.status_code ==200 : 

    soup = BeautifulSoup(res.content,'html.parser')
    items = soup.select('div.list_item')
    index = 0 

    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    header = ['index', 'Title','Reply_CNT']
    excel_sheet.column_dimensions['B'].width = 50
    excel_sheet.column_dimensions['C'].width = 15   
    excel_sheet.append(header)

    for item in items[:8] : #상위 8개 글만 갖고 오기 
        
        title = item.select_one('span.subject_fixed')
        reply_cnt = item.select_one('span.rSymph05')
        link = item.select_one('a.list_subject')
        #<a class="list_subject" href="news/18129728%EF%B9%96od=T31&amp;po=0&amp;category=0&amp;groupCd=.html" data-role="cut-string"> 
        
        if title : 
            index += 1
            # 인덱스를 3자리로 고정 (최대 자릿수만큼 0을 채우거나 공백을 채움)
            data1 = [index, title.get_text().strip(), '['+ reply_cnt.get_text()+']' ]
            excel_sheet.append(data1)
            
            res_reply = requests.get('https://davelee-fun.github.io/trial/board/'+link["href"])
            soup_reply = BeautifulSoup(res_reply.content, 'html.parser')
            replies = soup_reply.select('div.comment_content')

            index_reply = 0

            for reply in replies[:3] :
                index_reply += 1
                data2 = [f'{index}의 댓글 {index_reply}','', reply.get_text().strip().replace("\n", "").replace("\t", "")]
                excel_sheet.append(data2)
    
    excel_file.save('00.Title_and_Reply2.xlsx')
    excel_file.close()

else:
    print(f"Failed to retrieve the page. Status code: {res.status_code}")

```

7단계: 자동 줄바꿈과 볼드체 폰트 적용하기

```python
#자동 줄바꿈과 볼드체 폰트 적용하기

import requests, openpyxl
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font

#기본 설정
wrap_alignment = Alignment(wrap_text=True)
bold_font = Font(bold = True)
header_font = Font(bold = True, size = 15)
center_alignment = Alignment(horizontal='center', vertical='center')  # 가운데 정렬 설정

#엑셀 열기 
excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

#헤더 지정해주기 
header = ['index', 'Title','Reply_CNT']
excel_sheet.append(header)
for cell in excel_sheet[1]: #첫번째 행이 지정됨 
    cell.font = header_font #첫번째 줄을 볼드체로 
   
excel_sheet.column_dimensions['A'].width = 20
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 80

#크롤링 시작 
res  = requests.get('https://davelee-fun.github.io/trial/board/news.html')

if res.status_code ==200 : 

    soup = BeautifulSoup(res.content,'html.parser')
    items = soup.select('div.list_item')
    index = 0 
    row = 2  # row 변수를 루프 밖에서 초기화

    for item in items[:8] : #상위 8개 글만 갖고 오기 
        
        title = item.select_one('span.subject_fixed')
        reply_cnt = item.select_one('span.rSymph05')
        link = item.select_one('a.list_subject')
        #<a class="list_subject" href="news/18129728%EF%B9%96od=T31&amp;po=0&amp;category=0&amp;groupCd=.html" data-role="cut-string"> 
        
        if title : 

            index += 1
            
            data1 = [index, title.get_text().strip(), '['+ reply_cnt.get_text()+']' ]
            excel_sheet.append(data1)
            
            # 방금 추가한 데이터에 대해 스타일 적용
            current_row = excel_sheet.max_row  # 방금 추가된 행의 인덱스를 가져옴
            for cell in excel_sheet[current_row]: # 자 행의 번호를 넣으면 ? 
                cell.alignment = wrap_alignment #게시글 자동 줄바꿈 
                cell.font = bold_font #게시글 볼드체 
            #row += 1

            excel_sheet[f'A{current_row}'].alignment = center_alignment #방금 추가한 행에 대해 A열 가운데 설정

            res_reply = requests.get('https://davelee-fun.github.io/trial/board/'+link["href"])
            soup_reply = BeautifulSoup(res_reply.content, 'html.parser')
            replies = soup_reply.select('div.comment_content')

            index_reply = 0

            for reply in replies[:3] :
                index_reply += 1
                data2 = [f'{index}의 댓글 {index_reply}','', reply.get_text().strip().replace("\n", "").replace("\t", "")]
                excel_sheet.append(data2)

                current_row = excel_sheet.max_row
                for cell in excel_sheet [current_row]:
                    cell.alignment=wrap_alignment

                # A열 값들 가운데 정렬
                excel_sheet[f'A{current_row}'].alignment = center_alignment
        
            row += 1
    
    excel_file.save('00.Title_and_Reply3.xlsx')
    excel_file.close()

else:
    print(f"Failed to retrieve the page. Status code: {res.status_code}")
```
